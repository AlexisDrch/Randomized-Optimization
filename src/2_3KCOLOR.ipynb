```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2^100 solutions possible : space
def clean_df(df):
    df['time'] = pd.to_numeric(df['time']).multiply(1000)
    df['fitness'] = pd.to_numeric(df['fitness']).multiply(1000)
    return df
    
```

```python
#RHC
rhc1_df = clean_df(pd.read_csv('./output/TWOCOL/TWOCOL_RHC_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

#SA
sa75_df = clean_df(pd.read_csv('./output/TWOCOL/TWOCOL_SA0.75_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
sa95_df = clean_df(pd.read_csv('./output/TWOCOL/TWOCOL_SA0.95_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

#GA
ga1_df = clean_df(pd.read_csv('./output/TWOCOL/TWOCOL_GA100_10_10_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
ga2_df = clean_df(pd.read_csv('./output/TWOCOL/TWOCOL_GA100_30_30_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
ga3_df = clean_df(pd.read_csv('./output/TWOCOL/TWOCOL_GA100_50_50_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

#MiMic
mimic50_40_df = clean_df(pd.read_csv('./output/TWOCOL/100_WOCOL_MIMIC50_40_0.9_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
mimic100_40_df = clean_df(pd.read_csv('./output/TWOCOL/100_WOCOL_MIMIC100_40_0.9_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
mimic200_40_df = clean_df(pd.read_csv('./output/TWOCOL/100_WOCOL_MIMIC200_40_0.9_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

mimic200_10_df = clean_df(pd.read_csv('./output/TWOCOL/100_WOCOL_MIMIC200_10_0.9_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
mimic200_30_df = clean_df(pd.read_csv('./output/TWOCOL/100_WOCOL_MIMIC200_30_0.9_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
mimic200_60_df = clean_df(pd.read_csv('./output/TWOCOL/100_WOCOL_MIMIC200_40_0.9_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
mimic200_100_df = clean_df(pd.read_csv('./output/TWOCOL/100_WOCOL_MIMIC200_100_0.9_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
```

```python
#MIMIC compare to keep and num of sample

fig, ax = plt.subplots()

ax.plot(mimic50_40_df['time'], mimic50_40_df['fitness'], c = 'orange', label = '50 samples')
ax.plot(mimic100_40_df['time'], mimic100_40_df['fitness'], c = 'green', label = '100 samples')
ax.plot(mimic200_40_df['time'], mimic200_40_df['fitness'], c = 'red', label = '200 samples')

#ax.plot(ga1_df['time'], mimic_df['fitness'], c = 'purple', label = 'mimic')
ax.legend()
ax.set_xlabel("Time (ms)")
ax.set_ylabel("Fitnes (10^-3)")
#ax.set_xlim(xmax = 80)
plt.savefig('./output/2color-samples.png')
plt.show()

fig, ax = plt.subplots()

ax.plot(mimic200_10_df['time'], mimic200_10_df['fitness'], c = 'orange', label = '10 tokeep')
ax.plot(mimic200_30_df['time'], mimic200_30_df['fitness'], c = 'green', label = '30 tokeep')
ax.plot(mimic200_40_df['time'], mimic200_40_df['fitness'], c = 'red', label = '40 tokeep')
ax.plot(mimic200_100_df['time'], mimic200_100_df['fitness'], c = 'blue', label = '100 tokeep')

#ax.plot(ga1_df['time'], mimic_df['fitness'], c = 'purple', label = 'mimic')
ax.legend()
ax.set_xlabel("Time (ms)")
ax.set_ylabel("Fitnes (10^-3)")
#ax.set_xlim(xmax = 80)
plt.savefig('./output/2color-tokeep.png')
plt.show()
```

```python
fig, ax = plt.subplots()


ax.plot(gaa10_30_df['time'], gaa10_30_df['fitness'], c = 'orange', label = 'ga')
ax.plot(rhc10_df['time'], rhc10_df['fitness'], c = 'green', label = 'rhc')
ax.plot(sa10_75_df['time'], sa10_75_df['fitness'], c = 'red', label = 'sa')
ax.plot(mimic200_40_df['time'], mimic200_40_df['fitness'], c = 'purple', label = 'mimic')
ax.legend()
ax.set_xlabel("Time (ms)")
ax.set_ylabel("Fitnes (10^-3)")
ax.set_xlim(xmin = 0, xmax= 1500)
plt.savefig('./output/2colors.png')
plt.show()
```
