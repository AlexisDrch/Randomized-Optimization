```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#backprop
res_backprop = pd.read_csv('./output/NN/backprop/backprop_log.csv')
res1_backprop = pd.read_csv('./output/NN/backprop/1_backprop_log.csv')
res2_backprop = pd.read_csv('./output/NN/backprop/2_backprop_log.csv')
res3_backprop = pd.read_csv('./output/NN/backprop/3_backprop_log.csv')
res4_backprop = pd.read_csv('./output/NN/backprop/4_backprop_log.csv')

res_m_backprop = (res_backprop + res1_backprop + res2_backprop + res3_backprop + res4_backprop) /5 

def find_best_score(df):
    best_row = df.loc[:,'acc_test'].values.argmax(axis = 0)
    return df.loc[best_row]

print(find_best_score(res1_backprop))
```

```python
res_m_backprop.head(n=5)
```

```python
fig, ax = plt.subplots()

ax.plot(res_backprop['elapsed'], res_m_backprop['acc_test'], c = 'blue', label = 'Test accuracy')
ax.plot(res_backprop['elapsed'], res_m_backprop['acc_train'], c = 'red', label = 'Train accuracy')
ax.legend()
plt.savefig('./output/backprop-ref.png')
plt.show()
```

<u> Analysis </u> : 
    
Ref : global max : 0.110 / time : 2.7 sec

```python
#RHC
res_rhc = pd.read_csv('./output/NN/rhc/rhc_log.csv')
res1_rhc = pd.read_csv('./output/NN/rhc/1_rhc_log.csv')
res2_rhc = pd.read_csv('./output/NN/rhc/2_rhc_log.csv')
res3_rhc = pd.read_csv('./output/NN/rhc/3_rhc_log.csv')
res4_rhc = pd.read_csv('./output/NN/rhc/4_rhc_log.csv')

res_m_rhc = (res_rhc + res1_rhc + res2_rhc + res3_rhc + res4_rhc ) / 5 
fig, ax = plt.subplots()

ax.plot(res_m_rhc['elapsed'], res_m_rhc['acc_train'], c = 'blue', label = 'Train Accuracy')
ax.plot(res_m_rhc['elapsed'], res_m_rhc['acc_test'], c = 'red', label = 'Test Accuracy')
ax.legend()
plt.show()

print(find_best_score(res_m_rhc))
```
