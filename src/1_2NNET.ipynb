```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


def find_best_score(df):
    best_row = df.loc[:,'acc_tst'].values.argmax(axis = 0)
    return df.loc[best_row]
```

```python
# GA

res_ga_300_20_20 = pd.read_csv('./output/NN/GA2/0_GA_300_20_20_LOG.csv')
res1_ga_300_20_20 = pd.read_csv('./output/NN/GA2/1_GA_300_20_20_LOG.csv')
res2_ga_300_20_20 = pd.read_csv('./output/NN/GA2/2_GA_300_20_20_LOG.csv')
res3_ga_300_20_20 = pd.read_csv('./output/NN/GA2/3_GA_300_20_20_LOG.csv')
res4_ga_300_20_20 = pd.read_csv('./output/NN/GA2/4_GA_300_20_20_LOG.csv')

res_ga_400_20_30 = pd.read_csv('./output/NN/GA2/0_GA_400_20_30_LOG.csv')
res1_ga_400_20_30 = pd.read_csv('./output/NN/GA2/1_GA_400_20_30_LOG.csv')
res2_ga_400_20_30 = pd.read_csv('./output/NN/GA2/2_GA_400_20_30_LOG.csv')
res3_ga_400_20_30 = pd.read_csv('./output/NN/GA2/3_GA_400_20_30_LOG.csv')
res4_ga_400_20_30 = pd.read_csv('./output/NN/GA2/4_GA_400_20_30_LOG.csv')

res_ga_500_20_40 = pd.read_csv('./output/NN/GA2/0_GA_500_20_40_LOG.csv')
res1_ga_500_20_40 = pd.read_csv('./output/NN/GA2/1_GA_500_20_40_LOG.csv')
res2_ga_500_20_40 = pd.read_csv('./output/NN/GA2/2_GA_500_20_40_LOG.csv')
res3_ga_500_20_40 = pd.read_csv('./output/NN/GA2/3_GA_500_20_40_LOG.csv')
res4_ga_500_20_40 = pd.read_csv('./output/NN/GA2/4_GA_500_20_40_LOG.csv')

res_m_ga_300_20_20 = (res_ga_300_20_20 + res1_ga_300_20_20 + res2_ga_300_20_20 + res3_ga_300_20_20 + res4_ga_300_20_20)/5
res_m_ga_400_20_30 = (res_ga_400_20_30 + res1_ga_400_20_30 + res2_ga_400_20_30 + res3_ga_400_20_30 + res4_ga_400_20_30)/5
res_m_ga_500_20_40 = (res_ga_500_20_40 + res1_ga_500_20_40 + res2_ga_500_20_40 + res3_ga_500_20_40 + res4_ga_500_20_40)/5



fig, ax = plt.subplots()

ax.plot(res_ga_300_20_20['elapsed'], res_ga_300_20_20['acc_tst'], c = 'green', label = 'p/cross/mutate : 300/20/20')
ax.plot(res_ga_400_20_30['elapsed'], res_ga_400_20_30['acc_tst'], c = 'blue', label = 'p/cross/mutate : 400/20/30')
ax.plot(res_ga_500_20_40['elapsed'], res_ga_500_20_40['acc_tst'],  c = 'red', label = 'p/cross/mutate : 500/20/40')


ax.set_xlabel("Time (s)")
ax.set_ylabel("Accuracy (%)")
ax.set_xlim(xmax = 80)
ax.legend()
plt.savefig('./output/3-ga-compare.png')
plt.show()

print(find_best_score(res_ga_300_20_20), find_best_score(res_ga_400_20_30), find_best_score(res_ga_500_20_40))
```
