```python
#

def clean_df(df):
    df['time'] = pd.to_numeric(df['time']).multiply(1000)
    #df['fitness'] = pd.to_numeric(df['fitness']).multiply(1000)
    return df
    
```

```python
# RHC
rhc1_df = clean_df(pd.read_csv('./output/TSalesman/TSP_RHC_1_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
rhc2_df = clean_df(pd.read_csv('./output/TSalesman/TSP_RHC_2_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
rhc3_df = clean_df(pd.read_csv('./output/TSalesman/TSP_RHC_3_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

#SA
sa_55_df = clean_df(pd.read_csv('./output/TSalesman/TSP_SA0.55_2_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

# GA 100 / 30 /30
ga1_df = clean_df(pd.read_csv('./output/TSalesman/TSP_GA100_10_10_1_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

# Mimic 
mimic_df = clean_df(pd.read_csv('./output/TSalesman/TSP_MIMIC100_50_0.9_5_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

rhc1_df.head(n=5)
```

```python
# GA 100 / 30 /30
ga1_df = clean_df(pd.read_csv('./output/TSalesman/TSP_GA100_10_10_1_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
ga2_df = clean_df(pd.read_csv('./output/TSalesman/TSP_GA100_30_30_1_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
ga3_df = clean_df(pd.read_csv('./output/TSalesman/TSP_GA100_50_50_1_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
```

```python
fig, ax = plt.subplots()


ax.plot(ga1_df['time'], ga1_df['fitness'], c = 'orange', label = '100/10/10')
ax.plot(ga2_df['time'], ga2_df['fitness'], c = 'red', label = '100/30/30')
ax.plot(ga3_df['time'], ga3_df['fitness'], c = 'blue', label = '100/50/50')

#ax.plot(ga1_df['time'], mimic_df['fitness'], c = 'purple', label = 'mimic')
ax.legend()
ax.set_xlabel("Time (ms)")
ax.set_ylabel("Fitnes (10^-3)")
#ax.set_xlim(xmax = 80)
plt.savefig('./output/tsp-ga.png')
plt.show()

#use ga 2 -> quicker and converge
```

```python
#RHC and SA
rhc10_df = clean_df(pd.read_csv('./output/TSalesman/TSP_RHC_2_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
sa2_75 = clean_df(pd.read_csv('./output/TSalesman/2_TSP_SA0.75_1_LOG.txt', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])

fig, ax = plt.subplots()

print(len(rhc10_df))
ax.plot(rhc10_df['time'], rhc10_df['fitness'], c = 'orange', label = 'rhc')
ax.plot(sa2_75['time'], sa2_75['fitness'], c = 'red', label = 'sa x0.75')
ax.legend()
ax.set_xlabel("Time (ms)")
ax.set_ylabel("Fitnes (10^-3)")
#ax.set_xlim(xmax = 80)
plt.show()

#better, but does not converge to the global max
```
