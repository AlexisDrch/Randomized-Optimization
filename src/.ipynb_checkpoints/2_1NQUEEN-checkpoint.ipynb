```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def clean_df(df):
    df['time'] = pd.to_numeric(df['time']).multiply(1000)
    df['iterations'] = pd.to_numeric(df['iterations']).multiply(1000)
    #df['fitness'] = pd.to_numeric(df['fitness']).multiply(1000)
    return df
    
```

```python
# RHC
rhc1_df = clean_df(pd.read_csv('./output/NQUEEN/NQUEEN_RHC_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
rhc2_df = pd.read_csv('./output/NQUEEN/NQUEEN_RHC_2_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,]
rhc3_df = pd.read_csv('./output/NQUEEN/NQUEEN_RHC_3_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,]


#SA
sa1_df = pd.read_csv('./output/NQUEEN/NQUEEN_SA0.75_1_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,]
sa2_df = clean_df(pd.read_csv('./output/NQUEEN/NQUEEN_SA0.75_2_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,])
sa3_df = pd.read_csv('./output/NQUEEN/NQUEEN_SA0.75_3_LOG.csv', names=['iterations', 'fitness', 'time', 'fevals'] ).loc[1:,]


# GA 100 / 30 /30
ga1_df = clean_df(pd.read_csv('./output/NQUEEN/NQUEEN_GA100_30_30_1_LOG.csv',names=['iterations', 'fitness', 'time', 'fevals']).loc[1:,])
ga2_df = pd.read_csv('./output/NQUEEN/NQUEEN_GA100_30_30_2_LOG.csv',names=['iterations', 'fitness', 'time', 'fevals']).loc[1:,]
ga3_df = pd.read_csv('./output/NQUEEN/NQUEEN_GA100_30_30_3_LOG.csv',names=['iterations', 'fitness', 'time', 'fevals']).loc[1:,]

# Mimic 
mimic1_df = pd.read_csv('./output/NQUEEN/NQUEEN_MIMIC200_10_0.7_1_LOG.csv',names=['iterations', 'fitness', 'time', 'fevals']).loc[1:,]
mimic2_df = clean_df(pd.read_csv('./output/NQUEEN/NQUEEN_MIMIC200_10_0.7_2_LOG.csv',names=['iterations', 'fitness', 'time', 'fevals']).loc[1:,])
mimic3_df = pd.read_csv('./output/NQUEEN/NQUEEN_MIMIC200_10_0.7_3_LOG.csv',names=['iterations', 'fitness', 'time', 'fevals']).loc[1:,]
```

```python
fig, ax = plt.subplots()


ax.plot(ga1_df['iterations'], ga1_df['fitness'], c = 'orange', label = 'ga')
ax.plot(sa2_df['iterations'], sa2_df['fitness'], c = 'red', label = 'sa')
ax.plot(rhc1_df['iterations'], rhc1_df['fitness'], c = 'blue', label = 'rhc')
ax.plot(mimic2_df['iterations'], mimic2_df['fitness'], c = 'purple', label = 'mimic')

#ax.plot(ga1_df['time'], mimic_df['fitness'], c = 'purple', label = 'mimic')
ax.legend()
ax.set_xlabel("Iteration (10^-3)")
ax.set_ylabel("Fitness")
ax.set_xlim(xmin = 0)
plt.savefig('./output/NQUEEN/nqueen-itera.png')
plt.show()
```

```python
fig, ax = plt.subplots()


ax.plot(ga1_df['iterations'], ga1_df['time'], c = 'orange', label = 'ga')
ax.plot(sa2_df['iterations'], sa2_df['time'], c = 'red', label = 'sa')
ax.plot(rhc1_df['iterations'], rhc1_df['time'], c = 'blue', label = 'rhc')
ax.plot(mimic2_df['iterations'], mimic2_df['time'], c = 'purple', label = 'mimic')

#ax.plot(ga1_df['time'], mimic_df['fitness'], c = 'purple', label = 'mimic')
ax.legend()
ax.set_xlabel("Iteration (10^-3)")
ax.set_ylabel("Time (ms)")
ax.set_xlim(xmin = 0)
plt.savefig('./output/NQUEEN/nqueen-times.png')
plt.show()
```
